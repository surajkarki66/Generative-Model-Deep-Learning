### Data Preparation

```python
def extract_data(filename, num_images):
    with gzip.open(filename) as bytestream:
        bytestream.read(16)
        buf = bytestream.read(28 * 28 * num_images)
        data = np.frombuffer(buf, dtype=np.uint8).astype(np.float32)
        data = data.reshape(num_images, 28,28)
        return data
```

```python
train_data = extract_data('dataset/train-images-idx3-ubyte.gz', 60000)
test_data = extract_data('dataset/t10k-images-idx3-ubyte.gz', 10000)
```

```python
def extract_labels(filename, num_images):
    with gzip.open(filename) as bytestream:
        bytestream.read(8)
        buf = bytestream.read(1 * num_images)
        labels = np.frombuffer(buf, dtype=np.uint8).astype(np.int64)
        return labels
```

```python
train_labels = extract_labels('dataset/train-labels-idx1-ubyte.gz',60000)
test_labels = extract_labels('dataset/t10k-labels-idx1-ubyte.gz',10000)
```

### Convolution Autoencoder

```python
batch_size = 128
epochs = 50
n_channels = 1
w, h = 28, 28
input_img = Input(shape = (w, h, n_channels))
```

#### Encoder

```python
def encoder(input_img):
    #input = 28 x 28 x 1 (wide and thin)
    conv1 = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img) #28 x 28 x 32
    pool1 = MaxPooling2D(pool_size=(2, 2))(conv1) #14 x 14 x 32
    conv2 = Conv2D(64, (3, 3), activation='relu', padding='same')(pool1) #14 x 14 x 64
    pool2 = MaxPooling2D(pool_size=(2, 2))(conv2) #7 x 7 x 64
    conv3 = Conv2D(128, (3, 3), activation='relu', padding='same')(pool2) #7 x 7 x 128 (small and thick)
    
    return conv3

```

#### Decoder

```python
def decoder(input_encoder):
    #decoder
    conv4 = Conv2D(128, (3, 3), activation='relu', padding='same')(input_encoder) #7 x 7 x 128
    up1 = UpSampling2D((2,2))(conv4) # 14 x 14 x 128
    conv5 = Conv2D(64, (3, 3), activation='relu', padding='same')(up1) # 14 x 14 x 64
    up2 = UpSampling2D((2,2))(conv5) # 28 x 28 x 64
    decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(up2) # 28 x 28 x 1
    return decoded
```

```python
def autoencoder(input_img):
    enc = encoder(input_img)
    dec = decoder(enc)
    model = Model(input_img, dec)
    return model
```

```python
model = autoencoder(input_img)
model.summary()
```

```python
model.compile(loss='mean_squared_error', optimizer = RMSprop())
```
