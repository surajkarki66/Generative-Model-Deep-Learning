### Imports

```python
import time
import os
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
img_path = './pokemon_pix2pix_dataset'
IMG_HEIGHT = 256
IMG_WIDTH = 256
BUFFER_SIZE = 400
BATCH_SIZE = 1
```

### Data Preparation

```python
def image_loader(image_file):
    image_read = tf.io.read_file(image_file)
    image = tf.image.decode_jpeg(image_read)
    dim = tf.shape(image)[1]
    dim = dim // 2
    input_img = image[:, :dim, :]
    original = image[:, dim:, :]
    input_img = tf.cast(input_img, tf.float32)
    original = tf.cast(original, tf.float32)
    return input_img, original
```

```python
input_img, original_img = image_loader(img_path+'/train/2.jpg')
```

```python
plt.figure()
plt.imshow(input_img/255.0)
plt.figure()
plt.imshow(original_img/255.0)
```

#### Data augmentation

```python
def resize_image(input_img, original, height, width):
    input_img = tf.image.resize(
        input_img,[height, width], method=tf.image.ResizeMethod.NEAREST_NEIGHBOR)
    original = tf.image.resize(
        original, [height, width], method=tf.image.ResizeMethod.NEAREST_NEIGHBOR)
    return input_img, original
```

```python
def cutout(input_img, original):
    stacked = tf.stack([input_img, original], axis=0)
    cut_image = tf.image.random_crop(
        stacked, size=[2, IMG_HEIGHT, IMG_WIDTH, 3])
    return cut_image[0], cut_image[1]
```

```python
def normalize(input_img, original):
    input_img = (input_img / 127.5) - 1
    original = (original / 127.5) - 1
    return input_img, original
```

```python
def add_noise(input_img, original):
    input_img,original = resize_image(input_img,original, 256, 256)
    input_img,original = cutout(input_img,original)
    if tf.random.uniform(()) > 0.5:
        input_img = tf.image.flip_left_right(input_img)
        original = tf.image.flip_left_right(original)
    return input_img,original
```

### Building Model

#### Generator

```python
def downsample(filters, size, apply_batchnorm=True):
    initializer = tf.random_normal_initializer(0., 0.02)
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Conv2D(filters, size, strides=2, padding='same',kernel_initializer=initializer, use_bias=False))
    if apply_batchnorm:
        model.add(tf.keras.layers.BatchNormalization())
    model.add(tf.keras.layers.LeakyReLU())
    return model
```

```python
def upsample(filters, size, apply_dropout=False):
    initializer = tf.random_normal_initializer(0., 0.02)
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Conv2DTranspose(filters, size, strides=2, padding='same',
                                              kernel_initializer=initializer,
                                              use_bias=False))
    model.add(tf.keras.layers.BatchNormalization())
    if apply_dropout:
        model.add(tf.keras.layers.Dropout(0.5))
    model.add(tf.keras.layers.ReLU())
    return model
```

```python
def generator_model():  # an U-net model
    down_stack = [
        downsample(64, 4, apply_batchnorm=False),
        downsample(128, 4),
        downsample(256, 4),
        downsample(512, 4),
        downsample(512, 4),
        downsample(512, 4),
        downsample(512, 4),
        downsample(512, 4),
    ]
    up_stack = [
        upsample(512, 4, apply_dropout=True),
        upsample(512, 4, apply_dropout=True),
        upsample(512, 4, apply_dropout=True),
        upsample(512, 4),
        upsample(256, 4),
        upsample(128, 4),
        upsample(64, 4),
    ]
    initializer = tf.random_normal_initializer(0., 0.02)
    last = tf.keras.layers.Conv2DTranspose(3, 4,
                                           strides=2,
                                           padding='same',
                                           kernel_initializer=initializer,
                                           activation='tanh')
    concat = tf.keras.layers.Concatenate()
    inputs = tf.keras.layers.Input(shape=[None, None, 3])
    x = inputs
    skips = []
    for down in down_stack:
        x = down(x)
        skips.append(x)
    skips = reversed(skips[:-1])
    for up, skip in zip(up_stack, skips):
        x = up(x)
        x = concat([x, skip])
    x = last(x)
    return tf.keras.Model(inputs=inputs, outputs=x)
```

```python
generator = generator_model()
```

```python
generator.summary()
```

### Discriminator

```python
def discriminator_model():
    initializer = tf.random_normal_initializer(0., 0.02)
    input_img = tf.keras.layers.Input(shape=[None, None, 3], name='input_image')
    target_img = tf.keras.layers.Input(shape=[None, None, 3], name='target_image')
    concat= tf.keras.layers.concatenate([input_img, target_img]) 
    layer1 = downsample(64, 4, False)(concat)
    layer2 = downsample(128, 4)(layer1) 
    layer3 = downsample(256, 4)(layer2) 
    pad1 = tf.keras.layers.ZeroPadding2D()(layer3) 
    conv = tf.keras.layers.Conv2D(512, 4, strides=1,
                                kernel_initializer=initializer,
                                use_bias=False)(pad1) 
    batchnorm1 = tf.keras.layers.BatchNormalization()(conv)
    leaky_relu = tf.keras.layers.LeakyReLU()(batchnorm1)
    pad2 = tf.keras.layers.ZeroPadding2D()(leaky_relu) 
    last = tf.keras.layers.Conv2D(1, 4, strides=1,
                                kernel_initializer=initializer)(pad2) 
    return tf.keras.Model(inputs=[input_img, target_img], outputs=last)
```

```python
discriminator = discriminator_model()
```

```python
discriminator.summary()
```

### Loss function

```python
alpha = 100
loss_object = tf.keras.losses.BinaryCrossentropy(from_logits=True)
```

#### Discriminator loss

```python
def discriminator_loss(disc_real_output, disc_generated_output):
    actual_loss = loss_object(tf.ones_like(disc_real_output), disc_real_output) # real loss
    generated_loss = loss_object(tf.zeros_like(disc_generated_output), disc_generated_output) # fake loss
    total = actual_loss + generated_loss
    return total
```

#### Generator loss

```python
def generator_loss(disc_generated_output, gen_output, target):
    gan_loss = loss_object(tf.ones_like(disc_generated_output), disc_generated_output)
    mean_error = tf.reduce_mean(tf.abs(target - gen_output))
    total = gan_loss + (alpha *mean_error)
    return total
```

### Training

```python
def load_training_data(image_file):
    input_data,original = image_loader(image_file)
    input_data,original = add_noise(input_data,original)
    input_data,original= normalize(input_data,original)
    return input_data,original
```

```python
train_image = tf.data.Dataset.list_files(img_path+'/train/*.jpg')
train_image = train_image.map(load_training_data,
                                  num_parallel_calls=tf.data.experimental.AUTOTUNE)
train_image= train_image.cache().shuffle(BUFFER_SIZE)
train_image= train_image.batch(1)
```

```python
def load_test_data(image_file):
    input_data,original = image_loader(image_file)
    input_data,original = add_noise(input_data,original)
    input_data,original= normalize(input_data,original)
    return input_data,original
```

```python
test_image = tf.data.Dataset.list_files(img_path+'/test/*.jpg')
test_image = test_image.map(load_test_data,
                                  num_parallel_calls=tf.data.experimental.AUTOTUNE)
test_image= test_image.cache().shuffle(BUFFER_SIZE)
test_image= test_image.batch(1)
```

```python
generator_optimizer = tf.keras.optimizers.Adam(2e-4, beta_1=0.5)
discriminator_optimizer = tf.keras.optimizers.Adam(2e-4, beta_1=0.5)
```

```python
@tf.function
def train_step(input_image, target):
    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        gen_output = generator(input_image, training=True)
        
        disc_real_output = discriminator([input_image, target], training=True)
        disc_generated_output = discriminator([input_image, gen_output], training=True)
        
        gen_loss = generator_loss(disc_generated_output, gen_output, target)
        disc_loss = discriminator_loss(disc_real_output, disc_generated_output)
        
    generator_gradients = gen_tape.gradient(gen_loss,
                                          generator.trainable_variables)
    discriminator_gradients = disc_tape.gradient(disc_loss,
                                               discriminator.trainable_variables)
    generator_optimizer.apply_gradients(zip(generator_gradients,
                                          generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(discriminator_gradients,
                                              discriminator.trainable_variables))
    return gen_loss, disc_loss
```

```python
def generate_images(model, test_input, tar):
    prediction = model(test_input, training=True)
    plt.figure(figsize=(15,15))
    display_list = [test_input[0], tar[0], prediction[0]]
    title = ['Input Image', 'ground_truth', 'Predicted Image']
    for i in range(3):
        plt.subplot(1, 3, i+1)
        plt.title(title[i])
        plt.imshow(display_list[i] * 0.5 + 0.5)
        plt.axis('off')
    plt.show()
```

```python
def train(train_ds, epochs, test_ds):
    for epoch in range(epochs):
        start = time.time()
        gen_loss_list = []
        disc_loss_list = []
        for input_image, target in train_ds:
            loss = train_step(input_image, target)
            gen_loss_list.append(loss[0])
            disc_loss_list.append(loss[1])
            
        g_loss = sum(gen_loss_list) / len(gen_loss_list)
        d_loss = sum(disc_loss_list) / len(disc_loss_list)
        
        clear_output(wait=True)
        print (f'Epoch {epoch+1}, gen loss={g_loss},disc loss={d_loss}')
        for example_input, example_target in test_ds.take(1):
            generate_images(generator, example_input, example_target)
        if (epoch + 1) % 20 == 0:
            print ('Time taken for epoch {} is {} sec\n'.format(epoch + 1,
                                                        time.time()-start))
```
